```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('salary.txt',header=None)
```

```python
df.rename(columns={0:'Salary'},inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.hist(df.Salary,bins=100)
ax.set_xlabel('Salary Range')
ax.set_ylabel('count/frequency')
ax.set_title('Salary Histogram');
```

```python
fig, ax = plt.subplots()
ax.hist(df.Salary, cumulative=True, histtype='step')
ax.set_xlabel('Salary Range')
ax.set_ylabel('count/frequency')
ax.set_title('Salary Histogram');
```
